```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Gapminder-like dataset
gapminder = pd.read_csv('gapminder_dataset.csv')

# Compute the correlation matrix using only numeric columns
numeric_gapminder = gapminder.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_gapminder.corr()

# Create a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Seaborn Heatmap of Numeric Features in Gapminder Dataset')
plt.show()
```
